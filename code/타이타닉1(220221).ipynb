#2. 데이터 전처리

##2.1 결측치 처리

Age, Cabin, Embarked에 결측치가 발생된 것을 알 수 있다. Age는 나이로 평균을 이용하여 결측치를 해결하고 나머지 두 feature는 N이라는 값을 넣어 결측치를 해결하시오.

```python
df_titanic.info()
```

```python
df_titanic['Age']
df_titanic['Age'].mean()
round(df_titanic['Age'].mean(), 0)
```

```python
df_titanic['Age'].fillna(round(df_titanic['Age'].mean(), 0))
df_titanic['Age']
# df_titanic['Age'].fillna(round(df_titanic['Age'].mean(), 0), inplace=True)
df_titanic['Age'] = df_titanic['Age'].fillna(round(df_titanic['Age'].mean(), 0))
df_titanic['Age']
```

```python
df_titanic['Cabin']
df_titanic['Cabin'].fillna('N')
df_titanic['Cabin'].fillna('N', inplace = True)
df_titanic['Cabin']
```

```python
df_titanic['Embarked']
df_titanic['Embarked'].fillna('N')
df_titanic['Embarked'].fillna('N', inplace=True)
df_titanic['Embarked']
```

##2.3 문자열 수치화

```python
from sklearn import preprocessing

features=['Sex', 'Cabin', 'Embarked']

for feature in features:
  # LabelEncoder는 문자를 수치화하는 함수
  le = preprocessing.LabelEncoder()
  le.fit(df_titanic[feature])
  df_titanic[feature] = le.transform(df_titanic[feature])

df_titanic.head()
```

```python
# df_titanic.drop(['Name', 'Ticket'], axis=1, inplace =True)
df_titanic.head()
```

#3. 모델 만들기

##3.1 훈련 / 검증 데이터 만들기

> Quiz. 생존에 따른 데이터를 분석할 것이다. 다음 조건에 맞게 훈련용 데이터와 테스트용 데이터를 추출하시오.

*   Label 정보는 Survived를 이용한다.
*   나머지 정보를 Data로 활용한다.
*   테스트 데이터는 30%를 사용한다.
*   random_State는 62로 고정한다.

```python
from sklearn.model_selection import train_test_split

# Label 정보는 Survived를 이용한다.
df_label = df_titanic['Survived']
# 나머지 정보를 Data로 활용한다.
df_data = df_titanic.drop(['Survived'], axis=1)
# 테스트 데이터는 30%를 사용한다.
# random_state는 62로 고정한다.
X_train, X_test, y_train, y_test = train_test_split(df_data, df_label, test_size=0.3, random_state=62)
```

##3.2 결정트리

```python
from sklearn.tree import DecisionTreeClassifier

# 모델 로드
dt_clf = DecisionTreeClassifier()
# 학습
dt_clf.fit(X_train, y_train)
# 예측
pred = dt_clf.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score

# 정확도
accuracy_score(y_test, pred)
```

##3.3 랜덤 포레스트

```python
from sklearn.ensemble import RandomForestClassifier

# 모델 로드
rf_clf = RandomForestClassifier()
# 학습
rf_clf.fit(X_train, y_train)
# 예측
pred = rf_clf.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score

# 정확도
accuracy_score(y_test, pred)
```

##3.4 로지스틱 회귀

```python
from sklearn.linear_model import LogisticRegression
# 모델 로드
lr = LogisticRegression(max_iter=1000)
# 학습
lr.fit(X_train, y_train)
# 예측
pred = lr.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
# 정확도
accuracy_score(y_test, pred)
```
