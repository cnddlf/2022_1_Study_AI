```python
# 배달 시간 = 거리 * w+b
import tensorflow as tf

# 배달 시간
label = 33
distance=850

# tensorflow 형식의 변수 지정
# 실수 0으로 지정
w=tf.Variable(0.)
b=tf.Variable(0.)
```

```python
# 손실 함수는 예측값과 실제값의 차이를 줄이기 위한 함수이다.
def loss_func():
  # 예측값 계산
  pred = distance * w + b
  # 실 결과인 label과 예측값 사이의 손실 정도를 계산하여 0과 가까워 질때 까지 반복
  return tf.square(pred-label)
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.1)

for i in range(1000):
  opt.minimize(loss_func, var_list=[w,b])
  print(w.numpy(), ' : ', b.numpy())
```

```python
# 위의 코드를 이용하여 최적의 w, b를 구하고 그 때까지의 횟수를 구하시오.
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.1)

w_back = tf.Variable(0.)
b_back = tf.Variable(0.)

cnt = tf.Variable(0.)

while True:
  opt.minimize(loss_func, var_list=[w,b])
  pred = distance*w+b

  if(w_back.numpy()==w.numpy()) and (b_back.numpy()==b.numpy()) or cnt.numpy()>1000:
    break
  w=tf.Variable(0.)
  b=tf.Variable(0.)

  cnt.assign_add(1)

print(cnt.numpy(), ' : ', w.numpy(), ' : ', b.numpy())
```
