```python
dataset=[
         ['과자', '라면', '썬크림', '달걀'],
         ['과자', '라면', '맥주'],
         ['라면', '소주', '달걀'],
         ['소주', '맥주', '과자'],
         ['라면', '파', '달걀']
]
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_matrix = te.fit_transform(dataset)
te_matrix
```

```python
import pandas as pd
df_sales = pd.DataFrame(te_matrix, columns = te.columns_)
df_sales
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df_sales, min_support=0.4, use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets['len']=frequent_itemsets['itemsets'].apply( lambda x : len(x))
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules

ar = association_rules(frequent_itemsets, min_threshold=0.3)
ar.sort_values('lift', ascending=False)
```

```python
dataset = [ 
           ['아메리카노', '카페라떼'], 
           ['카페라떼', '아메리카노', '카푸치노'], 
           ['바닐라라떼', '아메리카노'], 
           ['녹차라떼', '카페라떼', '아메리카노'], 
           ['카페모카', '아메리카노'], 
           ['아메리카노', '카페라떼'], 
           ['초콜릿', '아메리카노'], 
           ['아메리카노'], 
           ['카페모카', '카페라떼'] 
           ]
```

```python
def getDataFrame(dataset):
  te = TransactionEncoder()
  te_matrix = te.fit_transform(dataset)
  return pd.DataFrame(te_matrix, columns = te.columns_)
```

```python
df_data = getDataFrame(dataset)
df_data
```

```python
freq_items = apriori(df_data, min_support=0.1, use_colnames=True)
freq_items
```

```python
ar = association_rules(freq_items, min_threshold=0.3)
ar.sort_values('lift', ascending=False)
```
