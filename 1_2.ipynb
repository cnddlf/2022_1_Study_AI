```python
import numpy as np

df2 = df
#df2.replace(np.nan, 'test')

df2.replace({'이용거리':np.nan}, {'이용거리':'test'})
#df2.head
```

```python
# df[df['대여소번호']==646]['이용거리']
mean646 = df2[df2['대여소번호']==112]['이용거리'].mean()
mean646 = round(mean646, 0)
df2[df2['대여소번호']==112] = df2[df2['대여소번호']==112].replace({'이용거리':np.nan}, {'이용거리':mean646})
df2
```

```python
def ReplaceNaN(rentNum):
  mean646 = df2[df2['대여소번호']==rentNum]['이용거리'].mean()
  mean646 = round(mean646, 0)
  df2[df2['대여소번호']==112] = df2[df2['대여소번호']==112].replace({'이용거리':np.nan}, {'이용거리':mean646})
  return df2

df3 = ReplaceNaN(103)
df3
```

```python
for idx in [646, 112, 103]:
  mean646 = df2[df2['대여소번호']==idx]['이용거리'].mean()
  mean646 = round(mean646, 0)
  df2[df2['대여소번호']==112] = df2[df2['대여소번호']==112].replace({'이용거리':np.nan}, {'이용거리':mean646})
df2
```

```python
for idx in [646, 112, 103]:
  df3 = ReplaceNaN(idx)
df3
```

```python
lst = df2['대여소번호']

for idx in lst:
  df3 = ReplaceNaN(idx)

df3.isnull().sum()
```
